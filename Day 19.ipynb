```python
import copy

test = """px{a<2006:qkq,m>2090:A,rfg}
pv{a>1716:R,A}
lnx{m>1548:A,A}
rfg{s<537:gd,x>2440:R,A}
qs{s>3448:A,lnx}
qkq{x<1416:A,crn}
crn{x>2662:A,R}
in{s<1351:px,qqz}
qqz{s>2770:qs,m<1801:hdj,R}
gd{a>3333:R,R}
hdj{m>838:A,pv}

{x=787,m=2655,a=1222,s=2876}
{x=1679,m=44,a=2067,s=496}
{x=2036,m=264,a=79,s=2244}
{x=2461,m=1339,a=466,s=291}
{x=2127,m=1623,a=2188,s=1013}""".split(
    "\n"
)

source = test

workflows = {}
paths = []

for l, workflow in enumerate(source):
    if workflow == "":
        break

    name, rule_list = workflow.split("{")
    rules = rule_list[:-1].split(",")

    workflows[name] = rules


def update_ranges(op, n, c, _ranges):
    new_ranges = copy.deepcopy(_ranges)

    if op == "<":
        if int(c) < new_ranges[n][1]:
            new_ranges[n][1] = int(c) - 1
    if op == ">":
        if int(c) > new_ranges[n][0]:
            new_ranges[n][0] = int(c) + 1

    return new_ranges


def score_range(_ranges):
    iscore = 1
    for k, v in _ranges.items():
        if v[0] > v[1]:
            return 0
        iscore *= v[1] - v[0] + 1
    return iscore


totals = []


def perform_workflow(wname, ranges, conditions=[], nodes=[]):
    if wname == "A":
        st = score_range(ranges)
        totals.append(st)
        print(ranges, "A-base", st, nodes + [wname])
        paths.append(conditions)
        return True
    if wname == "R":
        return False

    r = 0
    rule = workflows[wname][r]
    inverses = []

    while ":" in rule:
        condition, action = rule.split(":")
        if ">" in condition:
            parts = condition.split(">")
            inverse = f"{parts[0]}<{int(parts[1])+1}"
        elif "<" in condition:
            parts = condition.split("<")
            inverse = f"{parts[0]}>{int(parts[1])-1}"
        else:
            assert False

        n = condition[0]
        op = condition[1]
        c = condition[2:]

        new_ranges = update_ranges(op, n, c, ranges)

        for inv_condition in inverses:
            n = inv_condition[0]
            op = inv_condition[1]
            c = inv_condition[2:]
            new_ranges = update_ranges(op, n, c, new_ranges)

        print(wname, rule, condition)

        perform_workflow(
            action, new_ranges, conditions + inverses + [condition], nodes + [wname]
        )
        inverses.append(inverse)

        r += 1
        rule = workflows[wname][r]

    inverse_ranges = copy.deepcopy(ranges)

    for condition in inverses:
        n = condition[0]
        op = condition[1]
        c = condition[2:]

        inverse_ranges = update_ranges(op, n, c, inverse_ranges)

    print(inverses)
    perform_workflow(rule, inverse_ranges, conditions + inverses, nodes + [wname])


perform_workflow("in", {"x": [1, 4000], "m": [1, 4000], "a": [1, 4000], "s": [1, 4000]})

# print(totals)
print(sum(totals))

print("---")
print(paths)

t = 0
for p in paths:
    ranges = {"x": [1, 4000], "m": [1, 4000], "a": [1, 4000], "s": [1, 4000]}

    for c in p:
        n = c[0]
        oper = c[1]
        val = int(c[2:])
        if oper == ">":
            ranges[n][0] = max(ranges[n][0], val + 1)
        else:
            ranges[n][1] = min(ranges[n][1], val - 1)

    s = score_range(ranges)
    print(ranges, s)
    t += s
print(t)
```
