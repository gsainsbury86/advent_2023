```python
import requests

year = 2023
day = 25

url = f"https://adventofcode.com/{year}/day/{day}/input"

with open("cookie.txt", "r") as file:
    session = file.read()

headers = {"cookie": f"session={session}"}

with open(f"inputs/day_{str(day).zfill(2)}_input.txt", "wb") as file:
    file.write(requests.get(url, headers=headers).content)

aoc = "".join(open(f"inputs/day_{str(day).zfill(2)}_input.txt", "r"))


def use_input(input_string, debug=False):
    source = [x.strip() for x in input_string.strip().split("\n")]
    if debug:
        print(source)
    G = [[x for x in line] for line in source]
    R = len(G)
    C = len(G[0])
    if debug:
        print(G[:2][:2], R, C)

    return source, G, R, C


lines, G, R, C = use_input(aoc, True)
```

```python
test = """jqt: rhn xhk nvd
rsh: frs pzl lsr
xhk: hfx
cmg: qnr nvd lhk bvb
rhn: xhk bvb hfx
bvb: xhk hfx
pzl: lsr hfx nvd
qnr: nvd
ntq: jqt hfx bvb xhk
nvd: lhk
lsr: lhk
rzs: qnr cmg lsr rsh
frs: qnr lhk lsr"""

lines, G, R, C = use_input(aoc, True)


import plotly.graph_objects as go
import networkx as nx

# Create a simple graph
G = nx.Graph()
# G.add_nodes_from([1, 2, 3, 4])
# G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4)])


nodes = set({})
edges = set({})

for line in lines:
    component, rhs = line.split(":")
    connections = rhs.split()
    print(component, connections)
    for connection in connections:
        # links.add((component, connection))
        nodes.add(component)
        nodes.add(connection)
        edges.add(tuple(sorted((component, connection))))

G.add_nodes_from(list(nodes))
G.add_edges_from(list(edges))


pos = nx.fruchterman_reingold_layout(G)
# Create Plotly figure
fig = go.Figure()

# Add nodes to the plot
for node in G.nodes:
    fig.add_trace(
        go.Scatter(
            x=[], y=[], mode="markers", hovertext=str(node), marker=dict(size=10)
        )
    )

# Add edges to the plot
for edge in G.edges:
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    fig.add_trace(
        go.Scatter(
            x=[x0, x1, None],
            y=[y0, y1, None],
            mode="lines",
            line=dict(width=1),
        )
    )

x0 = [pos[node][0] for node in G.nodes]
y0 = [pos[node][1] for node in G.nodes]
fig.add_trace(
    go.Scatter(
        x=x0,
        y=y0,
        mode="markers",
        marker=dict(
            symbol="circle-dot",
            size=5,
            color="#6959CD",
            line=dict(color="rgb(50,50,50)", width=0.5),
        ),
        text=[str(node) for node in G.nodes],
        line=dict(width=1),
    )
)


# Update layout
fig.update_layout(showlegend=False, hovermode="closest")

# Show the plot
fig.show()
```
