```python
import requests

year = 2023
day = 11

url = f"https://adventofcode.com/{year}/day/{day}/input"

with open("cookie.txt", "r") as file:
    session = file.read()

headers = {"cookie": f"session={session}"}

with open(f"inputs/day_{str(day).zfill(2)}_input.txt", "wb") as file:
    file.write(requests.get(url, headers=headers).content)

lines = [x.strip() for x in open(f"inputs/day_{str(day).zfill(2)}_input.txt", "r")]
print(lines[:5])
```

```python
test = """...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#.....""".split(
    "\n"
)

source = lines

insert_hor = []
insert_vert = []

for i, line in enumerate(source):
    if not "#" in line:
        insert_hor.append(i)

for i in range(0, len(source[0])):
    has_gal = False
    for j in range(0, len(source)):
        if source[j][i] == "#":
            has_gal = True

    if not has_gal:
        insert_vert.append(i)

print(insert_hor)
print(insert_vert)

map_ = []
off = 0
for i in range(0, len(source)):
    map_.append(source[i])
    if i in insert_hor:
        map_.append("".join(["X"] * len(source[0])))
        off += 1

for m in map_:
    print(m)

for j in range(0, len(map_)):
    off = 0
    for x in insert_vert:
        map_[j] = map_[j][: x + off] + "X" + map_[j][x + off :]
        off += 1

for m in map_:
    print(m)
```

```python
print(len(map_))
print(len(map_[0]))

galaxies = []
for r, row in enumerate(map_):
    for c, col in enumerate(row):
        if col == "#":
            galaxies.append((r, c))

import itertools

combos = list(itertools.combinations(galaxies, 2))


def dist(combo):
    return abs(combo[0][0] - combo[1][0]) + abs(combo[0][1] - combo[1][1])


BIG_DIST = 1000000


def dist_2(combo):
    start_c = combo[0][0]
    start_r = combo[0][1]

    end_c = combo[1][0]
    end_r = combo[1][1]

    distance = 0
    for n in range(min(start_c, end_c), max(start_c, end_c)):
        if map_[n][start_r] == "." or map_[n][start_r] == "#":
            distance += 1
        elif map_[n][start_r] == "X":
            distance += BIG_DIST - 1
        else:
            pass

    for m in range(min(start_r, end_r), max(start_r, end_r)):
        if map_[start_c][m] == "." or map_[start_c][m] == "#":
            distance += 1
        elif map_[start_c][m] == "X":
            distance += BIG_DIST - 1
        else:
            pass
        # print(map_[j][i])

    return distance


sums = []
for combo in combos:
    sums.append(dist_2(combo))

print(sum(sums))
```
